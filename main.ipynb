- Empezamos a hacer un merge de las columnas que esténn repetidas, para tener un solo Tablón con toda las columnas necesarias para poder hacer una visualizacion de datos de manera correcta y limpia.

```python
merge_1= pd.merge(diamonds_city, diamonds_transactional, on= 'city_id')
merge_1
```

2.- merge diamonds_clarity y diamonds_properties en su columna clarity ID

```python
merge_2= pd.merge(diamonds_clarity, diamonds_properties,on='clarity_id')
merge_2
```

3.- merge diamonds_color y diamondos_properties en su columna color ID

```python
merge_3= pd.merge(diamonds_color, diamonds_properties, on= 'color_id')
merge_3
```

```python
merge_4 = pd.merge(diamonds_cut, diamonds_properties,on='cut_id')
merge_4
```

```python
merge_5=pd.merge(diamonds_dimensions, diamonds_properties, on='index_id')
merge_5
```

```python
merge_5_of= pd.merge(merge_5, diamonds_transactional, on='index_id')
merge_5_of
```

6.- vamos a merge nuestros dataframes into 1 (Tablon)

```python
merged_df = pd.merge(merge_1, merge_2, on='index_id')
merged_df
```

```python
merged_df = pd.merge(merged_df, merge_3, on=['clarity_id','index_id','cut_id','color_id'])
merged_df
```

```python
merged_df = pd.merge(merged_df, merge_4, on =['color_id','cut_id','index_id','clarity_id'])
merged_df
```

```python
merged_df = pd.merge(merged_df, merge_5, on =['cut_id','color_id','clarity_id','index_id'])
merged_df
```

```python
merged_df = merged_df.drop(['city_id','index_id','clarity_id','color_id','cut_id'],axis=1)
merged_df
```
